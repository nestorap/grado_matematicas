## Ejercicio 2
Utiliza las sumas de Riemann para aproximar el área delimitada por la gráfica de las funciones. Divide el intervalo dado en subintervalos idénticos y calcula las sumas de Riemann utilizando el punto medio de los subintervalos para.

```python
def f_de_x(x):
    return np.sqrt(1 - (np.abs(x) - 1) ** 2)

def g_de_x(x):
    return np.arccos(1 - np.abs(x)) - np.pi

def suma_riemann(f : callable, a: float, b: float, n: int) -> float:
    dx = (b - a) / n
    x = np.linspace(a, b, n, endpoint=False)  # Puntos de evaluación
    return np.sum(f(x) * dx)
```

```python
K = np.pi

nlist = [10, 20, 40]

for n in nlist:
    print(f"Suma de Riemann de la funcion f con n={n}: {suma_riemann(f=f_de_x, a=-2, b=2, n=n)}")
    print(f"Suma de Riemann de la funcion g con n={n}: {suma_riemann(f=g_de_x, a=-2, b=2, n=n)}")
```

```python
for n in nlist:
    print(f"Suma de Riemann de la funcion con n={n}: {suma_riemann(f=lambda x: f_de_x(x) - g_de_x(x), a=-2, b=2, n=n)}")
```
