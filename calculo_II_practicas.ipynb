```python
np.arcsin(-1)
np.arccos(-1)
```

```python
# Dibujamos la función arcoseno
# Valores de x (dominio de arccos: -1 a 1)
x = np.linspace(-1, 1, 100)

# Función arccos
y = np.arccos(x)

# Crear gráfica
plt.plot(x, y)

# Etiquetas
plt.title("Gráfica de arccos(x)")
plt.xlabel("x")
plt.ylabel("arccos(x)")

# Mostrar
plt.show()
```

```python
x = np.linspace(-2, 2, 100)
```

```python
# Guardamos la función f en una variable
f = np.sqrt(1 - (np.abs(x) - 1) ** 2)
# Graficamos
plt.figure()
plt.plot(x, f)
plt.title("f(x) = sqrt(1 - (|x|-1)^2)")
plt.grid()
plt.show()
```

```python
K = 3.14  # puedes cambiarlo luego
g = np.arccos(1 - np.abs(x)) - K

plt.figure()
plt.plot(x, g)
plt.title("g(x) = arccos(1 - |x|) - K")
plt.grid()
plt.show()
```

```python
K1 = np.pi
g1 = np.arccos(1 - np.abs(x)) - K1

# Graficamos un corazon
plt.figure()
plt.plot(x, f)
plt.plot(x, g1)
plt.title("Corazón")
plt.grid()
plt.show()
```
